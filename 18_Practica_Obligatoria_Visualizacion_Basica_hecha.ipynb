```python
df_ciudades = df_final.population_total.nlargest(5)
df_ciudades
```

```python
fig, ax = plt.subplots()

Cuidades = df_ciudades.index
Habitantes = df_ciudades
bar_colors = ['tab:blue', 'tab:grey', 'tab:red', 'tab:olive', 'tab:brown']

bar_container = ax.bar(Cuidades, Habitantes, color=bar_colors, alpha = 0.5)

ax.set_ylabel('Habitantes(miles)')
ax.set_xlabel('Ciudades')
ax.set_title('Cuidades más pobladas de California')


ax.bar_label(bar_container, fmt='{:,.0f}')


plt.show()
```

```python
#Histograma área total
fig, ax = plt.subplots()
plt.hist(df_final["area_total_km2"], bins = 50, color ="steelblue");
fontdict={'family': 'Bradley Hand itc', 'size': 18, 'color': 'black'}
plt.xlabel("Superficie (km2)", fontdict)
plt.ylabel("Numero de Ciudades", fontdict)
plt.legend(['Hist Area'], loc= 'upper right');
```

```python
datos= df_final['elevation_m']

colors = ('red')
plt.title('Diagrama de la Altitud de las ciudades de California')
plt.xlabel('Altitud(m)')
bplot = plt.boxplot(datos, vert= False,  whis= 1.5,  patch_artist= True)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color);
```

```python
# Creo la figura grande

fig, axes = plt.subplots(2, 2, figsize=(20,20) )

# Datos primera figura
Cuidades = df_ciudades.index
Habitantes = df_ciudades
bar_colors = ['tab:blue', 'tab:grey', 'tab:red', 'tab:olive', 'tab:brown']
bar_container = ax.bar(Cuidades, Habitantes, color=bar_colors, alpha = 0.5)

ax = axes[0, 0]
ax.bar(Cuidades, Habitantes, color=bar_colors, alpha = 0.5)
ax.set_ylabel('Habitantes(miles)')
ax.set_xlabel('Ciudades')
ax.set_title('Cuidades más pobladas de California')
ax.bar_label(bar_container, fmt='{:,.0f}')

#Datos segunda figura
df_final
ax = axes[0, 1]
ax.hist(df_final["area_total_km2"], bins = 50, color ="steelblue");
fontdict={'family': 'Bradley Hand itc', 'size': 18, 'color': 'black'}
ax.set_xlabel("Superficie (km2)", fontdict)
ax.set_ylabel("Numero de Ciudades", fontdict)
ax.legend(['Hist Area'], loc= 'upper right')

# Datos tercera figura
df_final
x = df_final['area_water_km2']
y = df_final['population_total']
ax = axes[1, 0]
ax.scatter(x, y, alpha=0.3, c='green')
ax.set_xlabel('Área de agua en la ciudad (km2)')
ax.set_ylabel('Población(miles de habitantes)')
ax.set_title('Población por kilómetro cuadrado de agua')

# Datos cuarta figura
datos= df_final['elevation_m']
colors = ('red')
ax = axes [1, 1]
bplot = ax.boxplot(datos, vert= False,  whis= 1.5,  patch_artist= True)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

ax.set_title('Diagrama de la Altitud de las ciudades de California')
ax.set_xlabel('Altitud(m)');
```
